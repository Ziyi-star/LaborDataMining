```python
import pandas as pd

# Load data from CSV file
data = pd.read_csv('Mouse.csv', header=None)
data
```

```python
# Create a copy of the data without the last two columns of data
data_copy = data.iloc[:, :-2].copy()
data_copy
```

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Set the number of clusters
c_values = [2, 3, 5]

# Perform clustering for each value of c
for c in c_values:
    # Create a k-means model with the number of clusters
    kmeans = KMeans(n_clusters=c, init='k-means++', n_init=1, max_iter=300, random_state=42)

    # Fit the model to the data
    kmeans.fit(data_copy)

    # Plot the data points and the clusters
    plt.scatter(data_copy.iloc[:, 0], data_copy.iloc[:, 1], c=kmeans.labels_)
    plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', color='r')
    plt.title(f'Clustering with {c} clusters')
    plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Load data from CSV file
data = pd.read_csv('Mouse.csv', header=None)

# Extract the labels from the second last column
encoder = LabelEncoder()
labels = encoder.fit_transform(data.iloc[:, -2])
labels
```

```python
# Extract the data without the last two columns
data_points = data.iloc[:, :-2].values

# Plot the data points with different colors according to the labels
plt.scatter(data_points[:, 0], data_points[:, 1], c=labels)
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
