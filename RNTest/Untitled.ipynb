```python
import matplotlib.pyplot as plt

# Definition der Bitsequenz für das Wort "Ablauf"
bit_sequence = "01100001011000100110110001100001011101100110110"

# Erstellen der x-Koordinaten für das NRZ-Signal
x = range(len(bit_sequence))

# Erstellen der y-Koordinaten für das NRZ-Signal
y = [int(bit) for bit in bit_sequence]

# Erstellen des Diagramms
plt.step(x, y, where='post')

# Beschriften der Achsen
plt.xlabel('Zeit')
plt.ylabel('Spannung')

# Festlegen der Achsenskalierung
plt.ylim(-0.5, 1.5)

# Anzeigen des Diagramms
plt.show()
```
